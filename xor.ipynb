```
import numpy as np

def Sigmoid(x):
    return 1/(1+np.exp(-x))

x=np.array([[0,0],
            [0,1],
            [1,0],
            [1,1]])

w1=np.array([[-2,  5,  4],
            [ 3,  6,  3]])
b1=np.array([ 2, -2, -5])

w2=np.array([[-4],
            [ 8],
            [-8]])

h=Sigmoid(np.dot(x,w1)+b1)
y=Sigmoid(np.dot(h,w2))

print (h)
print (y)
```
