```python
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeRegressor

df = pd.DataFrame({
    "Height": [1.6,1.6,1.5,1.8, 1.5, 1.4],
    "Color":["Blue","Green","Blue","Red","Green","Blue"],
    "Gender":["M","F","F","M","M","F"],
    "Weight":[88,76,56,73,77,57]})
```

```python
df = pd.get_dummies(df)
```

```python
df
```

```python
x = df.loc[:,df.columns!='Weight']
y = df['Weight']
F0 = np.mean(y)
print(F0)
```

```python
r1 = y - F0
print(r1)
```

```python
tree1  = DecisionTreeRegressor (random_state = 433, max_depth = 1)
r1_fit = tree1.fit(x, r1) # 독립변수로 잔차를 모형화 한다.
gamma1  = r1_fit.predict(x)
print ('잔차 예측값', gamma1)
```

```python
eta = 0.1
F1 = F0 + eta * gamma1 
print ('0번째 예측값:{}'.format(F0))
print ('첫번째 예측값:{}'.format(F1))
```

```python
import copy
F_new = np.mean(y) #초기화 모형
tree = DecisionTreeRegressor(random_state=433, max_depth =1) #약한 학습기
eta = 0.1 #학습률
T = 100
models = []
for t in range(T): 
    print(str(t+1)+'번째 업데이트')
    F_old = F_new
    r     = y - F_old        # 잔차
    r_fit = tree.fit(x,r)    # 잔차 적합
    gamma = r_fit.predict(x)
    models.append(copy.deepcopy(r_fit))
    print('잔차 예측값:{}'.format(np.round(gamma,2)))
    
    F_new = F_old + eta * gamma # 새로운 예측값
    
    print('예측값:{}'.format(np.round(F_new,2)))
```

```python
x_new = x.loc[0:0,:]
```

```python
x_new['Height'] = 2.0
```

```python
x_new
```

```python
F_old = F0
for i in range(T):
    F_new = F_old + eta * models[i].predict(x_new)
    F_old = F_new
    print(F_new)
```
