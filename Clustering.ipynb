### 군집분석 (계층적 클러스러팅)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x1 = np.array([9,7,5,6,11,10,8,4,8,6,5,3,12,7,6,5])
x2 = np.array([8,5,3,4,9,6,5,12,7,4,2,10,8,9,4,8])
X = np.concatenate((x1.reshape(-1,1), x2.reshape(-1,1)), axis = 1)
X
```

```python
from scipy.cluster.hierarchy import linkage, dendrogram
link = linkage(X, method='complete')
plt.figure(figsize=(10,5))
dendrogram(link)
plt.show()
```

```python
from scipy.cluster.hierarchy import fcluster
predict = fcluster(link, 5, criterion='distance')
predict
```

```python
np.unique(predict, return_counts=True)
```

```python
def dendrogram_plot(P=predict):
  for i in np.unique(predict, return_counts=True)[0]:
    plt.scatter(X[P == i][:,0],X[P == i][:,1])
  plt.show()    
dendrogram_plot()
```

```python
from scipy.cluster.hierarchy import fcluster
predict = fcluster(link, 4, criterion='distance')
predict
```

```python
np.unique(predict, return_counts=True)
```

```python
def dendrogram_plot(P=predict):
  for i in np.unique(predict, return_counts=True)[0]:
    plt.scatter(X[P == i][:,0],X[P == i][:,1])
  plt.show()    
dendrogram_plot()
```

### K-means 비계층적 군집분석

```python
x1 = np.array([9,7,5,6,11,10,8,4,8,6,5,3,12,7,6,5])
x2 = np.array([8,5,3,4,9,6,5,12,7,4,2,10,8,9,4,8])
X = np.concatenate((x1.reshape(-1,1), x2.reshape(-1,1)), axis = 1)
X
```

```python
from sklearn.cluster import KMeans
KM = KMeans(n_clusters = 3, init='k-means++',random_state=42)
KM.fit(X)
predict = KM.labels_
predict
```

```python
KM.cluster_centers_
```

```python
def cluster_plot(P=predict):
  for i in np.unique(predict, return_counts=True)[0]:
    plt.scatter(X[P == i][:,0],X[P == i][:,1])
cluster_plot()
plt.scatter(KM.cluster_centers_[0,0], KM.cluster_centers_[0,1], c='red')
plt.scatter(KM.cluster_centers_[1,0], KM.cluster_centers_[1,1], c='red')
plt.scatter(KM.cluster_centers_[2,0], KM.cluster_centers_[2,1], c='red')
plt.show()
```

```python
print(KM.transform(X[0:1]))
KM.predict(X[0:1])
```

```python
print(KM.transform([[8,8]]))
KM.predict([[8,8]])
```

### iris datasets

```python
from sklearn.datasets import load_iris
iris = load_iris()
iris.data.shape, iris.target.shape
```

```python
from sklearn.cluster import KMeans
KM = KMeans(n_clusters = 3, init='k-means++', random_state=42 )
KM.fit(iris.data)
predict = KM.labels_
predict
```

```python
np.unique(predict, return_counts=True)
```

```python
#KM.cluster_centers_
print(KM.transform([[5,3,2,1]]))
KM.predict([[5,3,2,1]])
```

```python
from sklearn.preprocessing import StandardScaler
SS = StandardScaler()
SS.fit(iris.data)
iris.data_scaled = SS.transform(iris.data)
iris.data_scaled.shape
```

```python
from sklearn.cluster import KMeans
KM = KMeans(n_clusters = 3, init='k-means++', random_state=42 )
KM.fit(iris.data_scaled)
predict = KM.labels_
predict
```

```python
np.unique(predict, return_counts=True)
```

```python
x =[[5,3,2,1]]
x_scaled = SS.transform(x)
print(KM.transform(x_scaled))
KM.predict(x_scaled)
```

### 최적의 K 탐색 - 엘보우 방법

```python
#n_clusters = 3 인 경우의 이너셔
KM.inertia_
```

```python
inertia = []
for i in range(2,10):
  KM = KMeans(n_clusters = i, init='k-means++', random_state=42 )
  KM.fit(iris.data_scaled)
  inertia.append(KM.inertia_)
plt.plot(range(2,10), inertia)
plt.show()
```
